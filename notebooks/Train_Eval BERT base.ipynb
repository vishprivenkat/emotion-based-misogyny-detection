# Reddit data preprocessing

```python
import pandas as pd
import html
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from cleantext import clean
import contractions
```

```python
|!pip install clean-text
!pip install contractions
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
df = pd.read_csv('./data/eval_videos.csv')
tweets = df['text']

for i in range (len(tweets)):
  #print(tweets[5])
  x = tweets[i].replace("\n","") #cleaning newline “\n” from the tweets
  x = tweets[i].replace("RT","") 
  tweets[i] = html.unescape(x)
tweets.head()
```

```python
sw = stopwords.words('english') #you can adjust the language as you desire
sw.remove('not') 
for i in range (len(tweets)):
  tweets[i] = re.sub('<[^>]+>', "", tweets[i])
  tweets[i] = contractions.fix(tweets[i])
  tweets[i]  = clean(tweets[i] , no_emoji=True) #emojis remove
  tweets[i] = re.sub(r"(@[A-Za-z0–9_]+)|[^\w\s]|#|http\S+", "", tweets[i])
  tweets[i] = re.sub("#[A-Za-z0-9_]+","", tweets[i]) #for hashtags
  tweets[i] = re.sub(r"http\S+", "", tweets[i]) #urls
  tweets[i] = re.sub(r"www.\S+", "", tweets[i])
  tweets[i] = tweets[i].lower()
  
  
tweets.head()
```

```python
df['full_text']=tweets
```

```python
tweets_to_token = tweets
sw = stopwords.words('english') #you can adjust the language as you desire
sw.remove('not') #we exclude not from the stopwords corpus since removing not from the text will change the context of the text

#Tokenize the tweets

for i in range(len(tweets_to_token)):
  tweets_to_token[i] = word_tokenize(tweets_to_token[i])

#Remove the Stopwords

for i in range(len(tweets_to_token)):
  tweets_to_token[i] = [word for word in tweets_to_token[i] if not word in sw]
tweets_to_token
```

```python
df= df.rename({'text': 'tokenised_text', 'full_text': 'text'}, axis=1) 
df
```

# Reddit data downsampling

```python
import pandas as pd 

dataset = pd.read_csv('./data/reddit_data/train.csv') 
```

```python
misogynistic, non_misogynistic = dataset[dataset['label']==1], dataset[dataset['label']==0] 
```

```python
from sklearn.utils import resample
non_misogynistic_downsample = resample(non_misogynistic,
             replace=True,
             n_samples=len(misogynistic),
             random_state=42)

print(non_misogynistic_downsample.shape)
```

```python
dataset = pd.concat([misogynistic, non_misogynistic]) 
dataset.to_csv('./data/reddit_data/train_downsampled.csv') 
```
