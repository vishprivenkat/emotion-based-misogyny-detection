```python
import plotly.graph_objects as go
import numpy as np

fig = go.Figure()
fig.add_trace(go.Box(y=df_train['valence'],x=df_train['label'], name = 'Valence',notched=True,
                marker_color = 'lightseagreen'))
fig.add_trace(go.Box(y=df_train['arousal'], x=df_train['label'],name='Arousal',notched=True,
                marker_color = 'indianred'))

fig.add_trace(go.Box(y=df_train['dominance'],x=df_train['label'], name = 'Dominance',notched=True,
                marker_color = 'blue'))
fig.update_layout(
    title='Comparison of Valence, Arousal and Dominance for Misogynistic vs Non-misogynistic data',
    #yaxis_title='normalized moisture',
    xaxis = dict(
        tickmode = 'array',
        tickvals = [0, 1],
        ticktext = ['Non-Misogynistic', 'Misogynistic']
    ),
    boxmode='group' # group together boxes of the different traces for each value of x
)
fig.show()
```
